```python
from abc import ABC, abstractmethod

# class num1
class Engine:
    def __init__(self, Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode= 'Normal_mode'):
        self.Code_Number = Code_Number
        self.Capacity = Capacity
        self.Number_of_cylinders = Number_of_cylinders
        self.Horsepower = Horsepower
        self._Mode = Mode        
    
    def Eco_mode(self):
        self._Mode = 'Eco_mode'
        return f"Mode Has Been Changed to {self._Mode} !"

    def Sport_mode(self):
        self._Mode = 'Sport_mode'
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Comfort_mode(self):
        self._Mode = 'Comfort_mode'
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Snow_mode(self):
        self._Mode = 'Snow_mode'
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Normal_mode(self):
        self._Mode = 'Normal_mode'
        return f"Mode Has Been Changed to {self._Mode} !"

    def Type_by_cylinder(self):
        return f"This is a V{self.Number_of_cylinders} Car !"

    
# class num2
class Cabin():
    def __init__(self, Material, Number_of_doors, Door_handle_type, Sunroof_status):
        self.Material = Material
        self.Number_of_doors = Number_of_doors
        self.Door_handle_type = Door_handle_type
        self.Sunroof_status = Sunroof_status
        self.__isdooropen = False
        self.__istrunkopen = False
        self.__ishoodopen = False
        
    def open_door(self):
        self.__isdooropen = True
        
    def open_trunk(self):
        self.__istrunkopen = True
        
    def open_hood(self):
        self.__ishoodopen = True
        
    def close_door(self):
        self.__isdooropen = False
        
    def close_trunk(self):
        self.__istrunkopen = False
        
    def close_hood(self):
        self.__ishoodopen = False
        
    def get_isdooropen(self):
        return self.__isdooropen
    
    def get_istrunkopen(self):
        return self.__istrunkopen
    
    def get_ishoodopen(self):
        return self.__ishoodopen
 
    
# class num3
class Available_cars(ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                 Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                 Sunroof_status, Gear_type):
        self.Model = Model 
        self.Color = Color
        self._Mileage = Mileage
        self.Date_of_manufacture = Date_of_manufacture
        self._Price = Price
        self.Number_of_seats = Number_of_seats 
        self.Zero_to_100 = Zero_to_100
        self._Seller = Seller
        self.__ison = False
        self.Engineobject = Engine(Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode)
        self.Cabinobject = Cabin(Material, Number_of_doors, Door_handle_type, Sunroof_status)
        self.Gear_type = Gear_type
        
    def start(self):
        self.__ison = True
        return "The Car is ON !"
    
    def shutdown(self):
        self.__ison = False
        return "The Car is OFF !"
        
    def get_iscaron(self):
        return self.__ison

    @abstractmethod
    def introduction(self):
        return "This is available !"
    
    @abstractmethod
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This Car has Manual Gear Type !" 

        
        
```

```python
# Class num4
class Sedan(Available_cars, ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                     Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                     Sunroof_status, Gear_type, Cabriolet):
        super().__init__(Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                         Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                         Sunroof_status, Gear_type)
        self.Cabriolet = Cabriolet
    
    @abstractmethod
    def introduction(self):
        return "This Car is an Sedan !"

    @abstractmethod
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This Sedan Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This Sedan Car has Manual Gear Type !"
   
    def Windshield_on(self):
        return "windshield is working !"
    
    def Windshield_off(self):
        return "windshield off !"


# Class num5
class SUV(Available_cars, ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                     Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                     Sunroof_status, Gear_type, Extra_seats, Off_road_tyre, Spare_tyre_placement):
        super().__init__(Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                         Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                         Sunroof_status, Gear_type)
        self.Extra_seats = Extra_seats
        self.Off_road_tyre = Off_road_tyre
    
    @abstractmethod
    def introduction(self):
        return "This Car is an SUV !"

    @abstractmethod
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This SUV Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This SUV Car has Manual Gear Type !"
 
    @abstractmethod
    def sparetyreplacement(self):
        if self.Spare_tyre_placement == "Inside the Trunk" :
            return "The Spare Tyre is Inside the Trunk of Car"
        elif self.Spare_tyre_placement == "On the Trunk door" :
            return "The Spare Tyre is on the Door of Car's Trunk"
   
    def extra_seats_status(self):
        if self.Extra_seats == True:
            return "This Car has Extra Seats !"
        else:
            return "This car Does not have Extra Seats !"
        
    def off_road_tyre_status(self):
        if self.Off_road_tyre == True:
            return "This Car is Equiped to Off_road Tyres !"
        else:
            return "This Car is not Equiped to Off_road Tyres !"
            
    
# Class num6
class Pickup(Available_cars, ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                     Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                     Sunroof_status, Gear_type, FourWD, Off_road):
        super().__init__(Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                         Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                         Sunroof_status, Gear_type)
        self.FourWD = FourWD
        self.Off_road = Off_road

    @abstractmethod
    def introduction(self):
        return "This Car is a Pickup !"

    @abstractmethod
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This Pickup Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This Pickup Car has Manual Gear Type !"
    
    def loaded(self):
        return "This Car is Loaded !"

    def not_loaded(self):
        return "This Car is not Loaded !"


# Class num7
class Supersport(Available_cars, ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                     Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                     Sunroof_status, Gear_type, Turbo, Spoiler):
        super().__init__(Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                         Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode,Material, Number_of_doors, Door_handle_type,
                         Sunroof_status, Gear_type)
        self.__Turbo = Turbo
        self.Spoiler = Spoiler
        self.__isturboon = False
  
    @abstractmethod
    def introduction(self):
        return "This Car is a Supersport !"

    @abstractmethod
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This Supersport Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This Supersport Car has Manual Gear Type !"
    
    def race_mode_on(self):
        self.__isturboon = True
        return "Race Mode is On and Turbo is about to Start !"
        
    def race_mode_off(self):
        self.__isturboon = False
        return "Race Mode is Off and Turbo is about to get Shutdown !"
    
    def get_Turbo(self):
        return self.__Turbo

    
    
    
```

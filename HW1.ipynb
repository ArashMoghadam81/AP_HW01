# **Creating 2 Composition for Parent Class, Parent Class, Abstraction & Encapsulation**

 * Engine and Cabin are here as Compositions in Parent Class
 * Class Available_cars is the Parent Class
 * We have Defined 2 Method as Abstraction
 * Here we have Called some Private and Protected  Attributes as Encapsulation
 #   
    

```python
from abc import ABC, abstractmethod

# class num1
class Engine:
    def __init__(self, Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode= 'Normal_mode'):
        self.Code_Number = Code_Number
        self.Capacity = Capacity
        self.Number_of_cylinders = Number_of_cylinders
        self.Horsepower = Horsepower
        self._Mode = Mode       #Encapsulation
    
    def Eco_mode(self):
        self._Mode = 'Eco_mode'         #Encapsulation
        return f"Mode Has Been Changed to {self._Mode} !"

    def Sport_mode(self):
        self._Mode = 'Sport_mode'       #Encapsulation
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Comfort_mode(self):
        self._Mode = 'Comfort_mode'         #Encapsulation
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Snow_mode(self):
        self._Mode = 'Snow_mode'        #Encapsulation
        return f"Mode Has Been Changed to {self._Mode} !"
    
    def Normal_mode(self):
        self._Mode = 'Normal_mode'      #Encapsulation
        return f"Mode Has Been Changed to {self._Mode} !"

    def Type_by_cylinder(self):
        return f"This is a V{self.Number_of_cylinders} Car !"

    
# class num2
class Cabin():
    def __init__(self, Material, Number_of_doors, Door_handle_type, Sunroof_status):
        self.Material = Material
        self.Number_of_doors = Number_of_doors
        self.Door_handle_type = Door_handle_type
        self.Sunroof_status = Sunroof_status
        self.__isdooropen = False       #Encapsulation
        self.__istrunkopen = False      #Encapsulation
        self.__ishoodopen = False       #Encapsulation
        
    def open_door(self):
        self.__isdooropen = True        #Encapsulation
        
    def open_trunk(self):
        self.__istrunkopen = True       #Encapsulation
        
    def open_hood(self):
        self.__ishoodopen = True        #Encapsulation
        
    def close_door(self):
        self.__isdooropen = False       #Encapsulation
        
    def close_trunk(self):
        self.__istrunkopen = False      #Encapsulation
        
    def close_hood(self):
        self.__ishoodopen = False       #Encapsulation
        
    def get_isdooropen(self):
        return self.__isdooropen        #Encapsulation
    
    def get_istrunkopen(self):
        return self.__istrunkopen       #Encapsulation
    
    def get_ishoodopen(self):
        return self.__ishoodopen        #Encapsulation
 
    
# class num3
class Available_cars(ABC):
    def __init__(self, Model, Color, Mileage, Date_of_manufacture, Price, Number_of_seats, Zero_to_100, Seller, 
                 Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode, Material, Number_of_doors, Door_handle_type,
                 Sunroof_status, Gear_type):
        self.Model = Model 
        self.Color = Color
        self._Mileage = Mileage         #Encapsulation
        self.Date_of_manufacture = Date_of_manufacture
        self._Price = Price         #Encapsulation
        self.Number_of_seats = Number_of_seats 
        self.Zero_to_100 = Zero_to_100
        self._Seller = Seller       #Encapsulation
        self.__ison = False         #Encapsulation
        self.Engineobject = Engine(Code_Number, Capacity, Number_of_cylinders, Horsepower, Mode)    #Composition
        self.Cabinobject = Cabin(Material, Number_of_doors, Door_handle_type, Sunroof_status)   #Composition
        self.Gear_type = Gear_type
        
    def start(self):
        self.__ison = True      #Encapsulation
        return "The Car is ON !"
    
    def shutdown(self):
        self.__ison = False         #Encapsulation
        return "The Car is OFF !"
        
    def get_iscaron(self):
        return self.__ison      #Encapsulation

    @abstractmethod         #Abstraction
    def introduction(self):
        return "This Car is available !"
    
    @abstractmethod         #Abstraction
    def get_gear_type(self):
        if self.Gear_type == "Automatic":
            return "This Car has Automated Gear Type !"
        elif self.Gear_type == "Manual":
            return "This Car has Manual Gear Type !" 
    
    
```

```python
# Class num16
class Exhibition:       #Aggregation
    def __init__(self, State, Address, Manager_name, Managing_age, Number_of_employees, Number_of_carslots, Cars_qty, Year, Annual_profit, Sold_cars= []):
        self.State = State
        self.Address = Address
        self.Manager_name = Manager_name
        self.Managing_age = Managing_age
        self.Number_of_employees = Number_of_employees
        self.Number_of_carslots = Number_of_carslots
        self.__Cars_qty = Cars_qty        #Encapsulation
        self.Year = Year
        self.__Annual_profit = Annual_profit        #Encapsulation
        self.__Sold_cars= Sold_cars        #Encapsulation
        self.Closed = False
        
    def get_number_of_available_cars(self):
        return self.__Cars_qty        #Encapsulation
    
    def Income_status(self):
        if self.__Annual_profit >= 10000000:        #Encapsulation
            return "This Exhibition Made a Profit That is Higher Than the Target !"
        elif self.__Annual_profit < 10000000:        #Encapsulation
            return "This Exhibition Made a Profit That is Lower Than the Target !"
        
    def promote_manager(self):
        if self.__Annual_profit >= 15000000 and self.Managing_age >= 4:        #Encapsulation
            self.Manager_name = "Exhibition Needs to Contract a New Manager !"
            self.Managing_age = 0
            return f"{self.Manager_name} has been Promoted and Will be Working at Headquarter Office !"
        else:
            return f"Sorry! {self.Manager_name} Can not be Promoted Due to Poor Performance or Lack of Exprience !"
        
    def close_exhibition(self):
        self.Closed = True
        return f"Exhibition {self.State} {self.Address} Would be closed for a While or Permanently. Please do Necessary Things in Order to Shutdown this Branch !"

    def exhibition_Status(self):
        if self.Closed == True :
            return f"Exhibition {self.State} {self.Address} has been closed for a While or Permanently !"
        else:
            return f"Exhibition {self.State} {self.Address} is Open and Working !"
 
    def soldcars_model(self):
        sold_models = []
        if len(self.__Sold_cars) == 0:
            return "This Branch haven't Sold any Car yet !"
        else:
            for sold in self.__Sold_cars:
                sold_models.append(sold.Model)
            return print(*sold_models , sep=', ') 
                       
    def display_exhibition(self):
        print("State: ", self.State)
        print("Address: ", self.Address)
        print("Manager_name: ", self.Manager_name)
        print("Managing_age: ", self.Managing_age)
        print("Number_of_employees: ", self.Number_of_employees)
        print("Number_of_carslots: ", self.Number_of_carslots)
        
    
```
